```python
ml.model = ml.model.str.replace(r'fsc-', r'').str.lower()
```

```python
colname_score = ml.columns[2]
colname_played = ml.columns[3]
colname_qual = ml.columns[4]
```

```python
grpd = ml.groupby(['lang'])
```

```python
for lang, df in grpd:
    # print(lang)
    display(df)
```

```python
models = grpd.get_group('de').model.tolist()
```

```python
def get_by_model(grpd, this_column):
    models = grpd.get_group('de').model.tolist()
    model2lang = {}
    #this_column = colname_played
    for model in models:
        lang2result = {}
        for lang, df in grpd:
            lang2result[lang] = (df[df.model == model][this_column].values[0])
            # print(lang, model, df[df.model == model][this_column].values[0])
        model2lang[model] = lang2result
    return pd.DataFrame(model2lang).transpose()
```

```python
played = get_by_model(grpd, colname_played)
score = get_by_model(grpd, colname_played)
quality = get_by_model(grpd, colname_qual)
```

could merge normal table and delta table, so that the delta is together with the actual value?

```python
left_df = played
right_df = played.subtract(played.en, axis=0)

def insert_delta(in_df, target_column):
    out_df = in_df.copy(deep=True)
    right_df = in_df.subtract(in_df.loc[:,target_column], axis=0)
    for index, row in out_df.iterrows():
        for column, value in row.iteritems():
            out_df.loc[index, column] = str(value) + ' (' + '{:.2f}'.format(right_df.loc[index, column]) + ')'
            #print(index, column, value)
    return out_df
```

```python
insert_delta(played, 'en')
```
